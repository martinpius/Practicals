#  String Formatting Techniques

```python
bus = "Shabiby"
route = "Dodoma - Dar"
fare = 45000

# Method 1: f-string (recommended)
print(f"{bus} travels on route {route} and charges {fare} TZS per passenger")

# Method 2: format()
print("{} travels on route {} and charges {} TZS per passenger".format(bus, route, fare))

# Method 3: old % formatting
print("%s travels on route %s and charges %d TZS per passenger" % (bus, route, fare))
```

# Identity vs Equality

```python
x = 1000
y = 1000

print("x == y :", x == y)       # value comparison
print("x is y :", x is y)       # identity comparison

a = [1, 2, 3]
b = [1, 2, 3]

print("a == b :", a == b)
print("a is b :", a is b)       # False: different memory locations
```

# More on Python data structure and control flow

# More on Dictionaries – Key-Value Pairs

```python
# Example: Dictionary for a bus info
bus_info = {
    "name": "Kili Express",
    "route": "Dar - Arusha",
    "capacity": 70,
    "air_conditioned": True
}

print("Bus Name:", bus_info["name"])
print("Route:", bus_info.get("route"))

# Update capacity
bus_info["capacity"] = 65
print("Updated Info:", bus_info)

# Add new key
bus_info["driver"] = "Musa"
```

# More on Sets – Unique & Unordered Collections

```python
# Example: Unique bus routes
routes = {"Dodoma-Moro", "Moro-Dar", "Dodoma-Mwanza", "Arusha-Dar"}
print("Unique Routes:", routes)

# Add a new route
routes.add("Arusha-Moshi")
print("Updated Routes:", routes)

# Membership test
print("Is 'Dodoma-Mwanza' in routes?", "Dodoma-Mwanza" in routes)
```

# More on If, Elif, Else (Control Logic)

```python
bus_speed = 95

if bus_speed > 80:
    print("Overspeeding! Penalty applies.")
elif 30 <= bus_speed <= 80:
    print("Safe speed")
else:
    print("Too slow for highway")
```
